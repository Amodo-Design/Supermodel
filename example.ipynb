First, import the relevant modules.

```python
import numpy as np
import matplotlib.pyplot as plt
from supermodel import modeling, config
```

Construct the case study that you want from `data/case_studies_vX.toml` and pass it to Supermodel.

```python
model = config.construct_case_study('Swiss courtroom')
model.verbose=True
results = modeling.Supermodel([(model,20)])

```

Do something with the results!

```python
fig, ax = plt.subplots(1, 2)

averaged = np.mean([result['Infected Occupants'] for result in results], axis=0)
stdev = np.std([result['Infected Occupants'] for result in results], axis=0)
ax[0].plot(averaged)
ax[0].fill_between(range(len(averaged)), np.clip(averaged-stdev,a_min=averaged[0], a_max = None), averaged+stdev, alpha=0.5)

for result in results:
    ax[1].plot(result['Infected Occupants'], alpha=0.2, color='black')

ax[0].set_title('Average infected occupants and stdev')
ax[1].set_title('Individual runs')
fig.supylabel('Infected occupants')
fig.supxlabel('Time [s]')
```
